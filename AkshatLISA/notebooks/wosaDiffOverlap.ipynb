```python
# Synthetic example: 10 Hz sine in white Gaussian noise
fs = 1000          # Hz
t  = np.arange(0, 5, 1/fs)
dt = t[1] - t[0]

nper = 256
nover_1 = (int)(0.25 * nper)
nover_2 = (int)(0.50 * nper)
nover_3 = (int)(0.75 * nper)
nover_4 = (int)(0.90 * nper)

sig = 0.5*np.sin(2*np.pi*10*t) + 0.2*np.random.randn(t.size)

psd_and_plot_hor(
    data = sig, 
    t = t, 
    dt = dt, 
    nper = 256, 
    title = "Sine signal", 
    window = 'hann', 
    scaling = 'density', 
    custom_wosa=True,
    average="mean",
    noverlap = nover_1
)
psd_and_plot_hor(
    data = sig, 
    t = t, 
    dt = dt, 
    nper = 256, 
    title = "Sine signal", 
    window = 'hann', 
    scaling = 'density', 
    custom_wosa=True,
    average="mean",
    noverlap = nover_2
)

psd_and_plot_hor(
    data = sig, 
    t = t, 
    dt = dt, 
    nper = 256, 
    title = "Sine signal", 
    window = 'hann', 
    scaling = 'density', 
    custom_wosa=True,
    average="mean",
    noverlap = nover_3
)

psd_and_plot_hor(
    data = sig, 
    t = t, 
    dt = dt, 
    nper = 256, 
    title = "Sine signal", 
    window = 'hann', 
    scaling = 'density', 
    custom_wosa=True,
    average="mean",
    noverlap = nover_4
)
```
